```python
#importing libraries
import pyttsx3
import speech_recognition as sr
import webbrowser
import datetime
import pywhatkit
import os
import yfinance as yf
import pyjokes
```

```python
#listen to microphone and return the audio as text with google
def transform():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        r.pause_threshold = 0.8
        said = r.listen(source)
        try:
            print('I am listening')
            q = r.recognize_google(said,language = 'en')
            return q
        except sr.Unknown.Value.Error:
            print("I didn't Understand")
            return 'i am waiting'
        except sr.RequestError:
            print("Sorry the service is down")
            return"I am waiting"
        except:
            return "i am waiting"
```

```python
#Greeting message
def speak(message):
    engine = pyttsx3.init()
    engine.say(message)
    engine.runAndWait()
```

```python
#return the week day name
def query_day():
    day = datetime.date.today()
#    print(day)
    weekday = day.weekday()
#    print(weekday)
    mapping = {
        0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday',6:'Sunday'
    }
    try:
        speak(f'Today is {mapping[weekday]}')
    except:
        pass
```

```python
#returns time
def query_time():
    time = datetime.datetime.now().strftime("%I:%M:%S")
    print(time)
    speak(f"Its {time[1]} o' clock and {time[3:5]} minutes")
```
